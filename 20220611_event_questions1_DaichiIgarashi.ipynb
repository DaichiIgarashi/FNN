問題１：FizzBuzz問題

```
for num in range(1, 101):
    string = ''

    # ここから記述

    # 15の倍数でFizzBuzzを出力
    if num % 15 == 0:
      print('FizzBuzz')

    # 3の倍数でFizzを出力
    elif num % 3 == 0: 
      print('Fizz')
    
    # 5の倍数でBuzzを出力
    elif num % 5 == 0:
      print('Buzz')
    
    
    
    # その他の場合数字を出力
    else:
      print(num)
    
    # ここまで記述

    print(string)
```
